#### Document-Term Metrix (문서 - 단어 행렬)

```python
dictionary = corpora.Dictionary(prePrecessedText)
print(dictionary)
```

#### 뽑아낸 데이터들은 bag of word (메트릭스) 형태로 만들어낸다 

```python
corpus = [dictionary.doc2bow(text) for text in prePrecessedText]
```

```python
print(corpus[0])
```

```python
print(dictionary[36])
```

#### TF-IDF 

```python
tfidf = models.TfidfModel(corpus)
tfidf_corpus = tfidf[corpus]
```

```python
tfidf_corpus[12]
```

#### models는 패키지내에 있는 라이브러리 

show_topic(n,x) n번째 기사의 상위 x개의 단어를 보겠다 

```python
model = models.ldamodel.LdaModel(corpus, num_topics=4, id2word=dictionary)
model.num_topics
```

```python
for id in range(model.num_topics):
    print('\n topic id : {}번째 기사'.format(id+1))
    #print('{}번째 기사 {} \n'.format(id, model.show_topic(id, 10)))
    for word, prob in model.show_topic(id):
        print('{}:{}'.format(word, prob))
```

### 텍스트 마이닝 시각화
