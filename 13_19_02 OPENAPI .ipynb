```python
import os
import sys
import json

import pandas as pd 

import urllib.request
import requests as req

from bs4 import BeautifulSoup as bs

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait

from fake_useragent import UserAgent
import time
```

```python
client_id = "3M4q5UEmMPvE8pMjI8lq"
client_secret = "v5aAcg6Tpi"
encText = urllib.parse.quote("코로나")

news_data = []

for idx in range(0,5):
    url = "https://openapi.naver.com/v1/search/news.json?query=" + encText + '&start=' + str(idx*10+1)# + '&display=100'  json 결과 

    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)lin
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        #print(response_body.decode('utf-8'))
        news_data.append(json.loads(response_body.decode('utf-8')))
    elselin
        print("Error Code:" + rescode)
```

```python
links = []
for pages in news_data:
    for item in pages['items']: 
        #print(item['link'])
        if 'https://news.naver.com/main' in item['link']:
            links.append(item['link'])
```

#### 크롬드라이브 옵션 켜기 

```python
# option 
options = webdriver.ChromeOptions()
options.add_argument("-headless")
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
browser = webdriver.Chrome('C:/Users/scien/Desktop/selfstudydata/Chrome_Driver/chromedriver.exe', options=options)
```

```python
# selenium
news_title = []
news_content = []

for url in links:    
    browser.get(url) #1
    response = browser.page_source
    soup = bs(response, 'html.parser') #2
    
    try: #5 
        title = soup.find('h3', id='articleTitle').get_text() #3
        news_title.append(title)
        
    except:#6
        continue 

        
    content = soup.find('div', id='articleBodyContents') #4
    news_content.append(''.join(content.find_all(text=True)).replace('\n', ''))
    
    time.sleep(2)
```
