```python
def cleantext(text):
                                # 띄어쓰기 필수!!!!!!!!!!!
    ret = re.sub('[^ㄱ-ㅎㅏ-ㅣ가-힣 ]', "", text)
    return ret
```

## 파이프라인 만들기 

```python
learner1 = Pipeline( [
    ('vect', CountVectorizer()),
    ('model', SVC(kernel='rbf'))
])

model = learner1
model.fit(train_x[:1000], train_y[:1000])
predict = model.predict(test_x[:1000])
print( np.mean(predict == test_y[:1000]))
```

```python
learner2 = Pipeline([
    ('vector', CountVectorizer()),
    ('model', SVC(kernel='linear'))
])

model = learner2
model.fit(train_x[:1000], train_y[:1000])
predict = model.predict(test_x[:1000])
print( np.mean(predict == test_y[:1000]))
```
