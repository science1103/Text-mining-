```python
with open('./nsmc_token_train.json', 'r', encoding='utf-8') as file:
    train_token_load = json.load(file)
```

```python
total_words = [word for line in train_token_load for word in line[0]]
```

```python
text = nltk.Text(total_words, name='NMSC')
pprint(text.vocab().most_common(10))
```
